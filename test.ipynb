```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import backend as k
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = tf.keras.models.load_model("Img_Classification.h5")
```

```python
def test_predict(model, x, classes):
    x = x.astype("float32")
    x /= 255.0
    x = np.expand_dims(x, axis = 0)
    prediction = model.predict(x, verbose = 0)
    print("Prediction: This is", classes[np.argmax(prediction)])
    plt.imshow(np.squeeze(x))
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
classes = ["bowser", "luigi", "mario", "pikachu", "sonic"]
for i in classes:
    path = "test_images/" + i + "_test.jpg"
    img = load_img(path, target_size = (128, 128), color_mode = "rgb")
    img_arr = img_to_array(img)
    test_predict(model, img_arr, classes)
```
